```python
import pandas as pd
import re
```

```python
#firstly important the file
data = pd.read_csv('fakeReviewData.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
#Creating dataframe
df = pd.DataFrame(data)
```

```python
#Removing duplicate by consdering the first occurence
temp = df.drop_duplicates(subset=None, keep='first', inplace=False)
df = pd.DataFrame(temp)
df
```

```python
#Checking up for missing value if.
missing_values = df.isna()
print(missing_values)
```

```python
#Convert all the data in dataset to lower order using str.lower()
df['category'] = df['category'].str.lower()
df['label'] = df['label'].str.lower()
df['text_'] = df['text_'].str.lower()
df
```

```python
# function for removing punctuation, special characters, and numbers.
def clean_text(text):
    if isinstance(text, str):  
        return re.sub(r'[^a-zA-Z\s]', '', text)  
    return text 
```

```python
#appyling it on text_ column
df['text_'] = df['text_'].apply(clean_text)
df
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')  
```

```python
# Tokenize each sentence using str.split()
df['Tokens'] = df['text_'].str.split()
df
```

```python
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from the tokens
def remove_stopwords(tokens):
    return [word for word in tokens if word.lower() not in stop_words]

# Apply the function on Tokens Column
df['Filtered Text'] = df['Tokens'].apply(remove_stopwords)
df[['text_', 'Filtered Text']]
```

```python
from nltk.stem import PorterStemmer 
```

```python
stemmer = PorterStemmer()
```

```python
#Stemming the filtered text
def apply_stemming(text):
    tokens = text.split()  
    return " ".join([stemmer.stem(word) for word in tokens])
```

```python
df['stemmed_text'] = df['text_'].apply(apply_stemming)
```

```python
df
```

```python
#Converting pre-processing data in numeric vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
X_bow = vectorizer.fit_transform(df['stemmed_text'])
```

```python
bow_df = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())
```
